```python
import os

base = "/content/gcp_yolo_dataset"

folders = [
    base + "/images/train",
    base + "/labels/train"
]

for f in folders:
    os.makedirs(f, exist_ok=True)

print("YOLO folders created")
```

```python
import os
import json
import shutil
import cv2
from tqdm import tqdm

# Dataset paths
train_root = "/content/drive/MyDrive/GCP_Assignment_Datasets/train_dataset"
json_path = os.path.join(train_root, "curated_gcp_marks.json")

# YOLO output folders
yolo_img_dir = "/content/gcp_yolo_dataset/images/train"
yolo_label_dir = "/content/gcp_yolo_dataset/labels/train"

os.makedirs(yolo_img_dir, exist_ok=True)
os.makedirs(yolo_label_dir, exist_ok=True)

# Load labels
with open(json_path) as f:
    data = json.load(f)

# Shape → class ID
shape_map = {
    "Cross": 0,
    "Square": 1,
    "L-Shaped": 2
}

# Bounding box size (pixels)
BOX_SIZE = 40

# Convert dataset
for rel_path, info in tqdm(data.items()):

    img_path = os.path.join(train_root, rel_path)

    if not os.path.exists(img_path):
        continue

    # Some records may not have shape label
    shape = info.get("verified_shape")
    if shape is None:
        continue

    img = cv2.imread(img_path)
    h, w = img.shape[:2]

    x = info["mark"]["x"]
    y = info["mark"]["y"]

    class_id = shape_map[shape]

    # Normalize coordinates
    x_center = x / w
    y_center = y / h
    bw = BOX_SIZE / w
    bh = BOX_SIZE / h

    # Create new image name
    new_name = rel_path.replace("/", "_")

    # Copy image to YOLO folder
    shutil.copy(img_path, os.path.join(yolo_img_dir, new_name))

    # Create label file
    label_file = os.path.join(yolo_label_dir, new_name.replace(".JPG", ".txt"))

    with open(label_file, "w") as f:
        f.write(f"{class_id} {x_center} {y_center} {bw} {bh}")
```

```python
config_text = """
path: /content/gcp_yolo_dataset
train: images/train

names:
  0: Cross
  1: Square
  2: L-Shaped
"""

with open("/content/gcp_yolo_dataset/gcp.yaml", "w") as f:
    f.write(config_text)

print("config file created")
```

```python
import os
import json
from ultralytics import YOLO

model = YOLO("/content/runs/detect/train2/weights/best.pt")

test_root = "/content/drive/MyDrive/GCP_Assignment_Datasets/test_dataset"

results_json = {}

shape_map = {0:"Cross",1:"Square",2:"L-Shaped"}

for root, dirs, files in os.walk(test_root):

    for f in files:

        if not f.lower().endswith(".jpg"):
            continue

        img_path = os.path.join(root, f)

        result = model(img_path)[0]

        if len(result.boxes)==0:
            continue

        box = result.boxes[0]

        x1,y1,x2,y2 = box.xyxy[0].tolist()

        x = (x1+x2)/2
        y = (y1+y2)/2

        cls = int(box.cls[0])

        rel_path = os.path.relpath(img_path, test_root)

        results_json[rel_path] = {
            "mark":{"x":x,"y":y},
            "verified_shape":shape_map[cls]
        }

with open("/content/predictions.json","w") as f:
    json.dump(results_json,f,indent=4)

print("predictions.json created")
```

```python
import os
import json
import math
from ultralytics import YOLO

# Load model
model = YOLO("/content/runs/detect/train2/weights/best.pt")

# Dataset paths
train_root = "/content/drive/MyDrive/GCP_Assignment_Datasets/train_dataset"
json_path = os.path.join(train_root, "curated_gcp_marks.json")

with open(json_path) as f:
    data = json.load(f)

shape_map = {
    0: "Cross",
    1: "Square",
    2: "L-Shaped"
}

total = 0
correct_class = 0
distance_errors = []

for rel_path, info in data.items():

    if "verified_shape" not in info:
        continue

    img_path = os.path.join(train_root, rel_path)

    if not os.path.exists(img_path):
        continue

    result = model(img_path)[0]

    if len(result.boxes) == 0:
        continue

    box = result.boxes[0]

    x1,y1,x2,y2 = box.xyxy[0].tolist()

    pred_x = (x1+x2)/2
    pred_y = (y1+y2)/2

    gt_x = info["mark"]["x"]
    gt_y = info["mark"]["y"]

    dist = math.sqrt((pred_x-gt_x)**2 + (pred_y-gt_y)**2)
    distance_errors.append(dist)

    pred_cls = shape_map[int(box.cls[0])]
    gt_cls = info["verified_shape"]

    if pred_cls == gt_cls:
        correct_class += 1

    total += 1


avg_distance = sum(distance_errors)/len(distance_errors)
class_accuracy = correct_class/total

print("Samples evaluated:", total)
print("Average localization error (pixels):", avg_distance)
print("Classification accuracy:", class_accuracy)
```
